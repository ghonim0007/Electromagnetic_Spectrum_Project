# Introduction to Remote Sensing and the Electromagnetic Spectrum

```python
# pip install numpy matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

##  Energy Calculation Using Wavelength
### 1- Electromagnetic wave equation:
####  c=vλ
##### Where c is the speed of light, v is the frequency, and λ is the wavelength.

### 2- Quantum energy:
####  Q=hv
##### Where h is Planck's constant, and v is the frequency.

### 3- Energy and wavelength:
#### Q=λ\hc​

```python
# الثوابت
h = 6.626e-34  # ثابت بلانك (J.s)
c = 3e8        # سرعة الضوء (m/s)

# نطاق الأطوال الموجية (من 0.4 ميكرومتر إلى 14 ميكرومتر)
wavelengths = np.linspace(0.4, 14, 1000)  # من الأشعة فوق البنفسجية إلى الأشعة تحت الحمراء

#  Q = hc / λ
energy = h * c / (wavelengths * 1e-6)  # تحويل الطول الموجي من ميكرومتر إلى متر

# رسم الطاقة مقابل الطول الموجي
plt.figure(figsize=(10, 6))
plt.plot(wavelengths, energy, label='Quantum Energy (Q)', color='blue')
plt.title("Quantum Energy vs Wavelength")
plt.xlabel("Wavelength (µm)")
plt.ylabel("Energy (Joules)")
plt.grid(True)
plt.legend()
plt.show()

# رسم الطيف الكهرومغناطيسي مع تصنيف الأجزاء المختلفة
plt.figure(figsize=(10, 6))
plt.plot(wavelengths, energy, label='Quantum Energy (Q)', color='blue')
plt.axvline(x=0.4, color='green', linestyle='--', label="UV Start")
plt.axvline(x=0.7, color='red', linestyle='--', label="Visible End")
plt.axvline(x=1.3, color='purple', linestyle='--', label="IR Start")

# إضافة تسمية للأجزاء
plt.fill_between(wavelengths, energy, where=(wavelengths >= 0.4) & (wavelengths <= 0.7), color='yellow', alpha=0.3, label='Visible')
plt.fill_between(wavelengths, energy, where=(wavelengths > 0.7) & (wavelengths <= 1.3), color='orange', alpha=0.3, label='Near IR')
plt.fill_between(wavelengths, energy, where=(wavelengths > 1.3) & (wavelengths <= 3), color='red', alpha=0.3, label='Mid IR')
plt.fill_between(wavelengths, energy, where=(wavelengths > 3), color='purple', alpha=0.3, label='Far IR')

# إعدادات الرسم
plt.title("Electromagnetic Spectrum with Energy and Wavelengths")
plt.xlabel("Wavelength (µm)")
plt.ylabel("Energy (Joules)")
plt.grid(True)
plt.legend()
plt.show()
```

### The relationship Q=h⋅v is used to calculate the energy of photons based on frequency.

#### Frequencies range from 1 GHz to 1 THz.
#### The graph shows the direct relationship between frequency and energy.

```python
frequencies = np.linspace(1e9, 1e12, 1000)

# Q = h * v 
energy_freq = h * frequencies

# رسم بياني للطاقة مقابل التردد
plt.figure(figsize=(10, 6))
plt.plot(frequencies, energy_freq, label='Quantum Energy (Q) as a function of frequency', color='green')
plt.title("Quantum Energy vs Frequency")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Energy (Joules)")
plt.grid(True)
plt.legend()
plt.show()
```

### Refraction curves of materials

#### Concept: Each material on the Earth's surface has a different response to the electromagnetic spectrum (reflection curves).
#### Plants: have a characteristic reflection in the red and near-infrared region.
#### Sand: shows a higher reflection in the middle spectrum.
#### Soil: has a lower reflectance curve overall.
##### Refraction curves are plotted for each material separately.

```python
# نطاق الأطوال الموجية (من 0.4 ميكرومتر إلى 2.5 ميكرومتر)
wavelengths = np.linspace(0.4, 2.5, 100)

# منحنيات الانكسار لمواد مختلفة (نباتات، رمل، تربة)
vegetation_reflectance = np.exp(-(wavelengths - 0.67)**2 / 0.1) + 0.2 * np.exp(-(wavelengths - 0.75)**2 / 0.1)
sand_reflectance = 0.6 + 0.2 * np.exp(-(wavelengths - 1.0)**2 / 0.2)
soil_reflectance = 0.4 + 0.3 * np.exp(-(wavelengths - 0.5)**2 / 0.3)

# رسم منحنيات الانكسار
plt.figure(figsize=(10, 6))
plt.plot(wavelengths, vegetation_reflectance, label="Vegetation", color="green")
plt.plot(wavelengths, sand_reflectance, label="Sand", color="orange")
plt.plot(wavelengths, soil_reflectance, label="Soil", color="brown")
plt.title("Spectral Reflectance Curves for Different Earth Surface Features")
plt.xlabel("Wavelength (µm)")
plt.ylabel("Reflectance")
plt.legend(loc="upper right")
plt.grid(True)
plt.show()
```
